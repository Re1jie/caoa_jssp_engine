```python
import pandas as pd
import numpy as np
```

```python
def tinjau_A_lj_terkecil(df, group_by=None):
    if 'A_lj' not in df.columns:
        raise ValueError("Kolom 'A_lj' tidak ditemukan dalam DataFrame yang diberikan.")
        
    if group_by:
        # Mencari indeks baris dengan nilai A_lj terkecil untuk setiap kelompok (misal per job per kapal)
        idx_min = df.groupby(group_by)['A_lj'].idxmin()
        result = df.loc[idx_min].sort_values('A_lj')
    else:
        # Mencari nilai A_lj terkecil secara keseluruhan
        min_val = df['A_lj'].min()
        result = df[df['A_lj'] == min_val]
        
    return result

# 1. Load dataset JSSP yang telah diproses
df_jssp = pd.read_csv("../data/processed/jssp_data.csv")

# 2. Tinjau nilai A_lj terkecil secara keseluruhan pada seluruh dataset
print("--- A_lj Terkecil Secara Keseluruhan ---")
display(tinjau_A_lj_terkecil(df_jssp))

# 3. Opsional: Tinjau nilai A_lj terkecil untuk setiap Job (Kapal)
print("\n--- A_lj Terkecil untuk Setiap Job ID ---")
display(tinjau_A_lj_terkecil(df_jssp, group_by='job_id'))
```
